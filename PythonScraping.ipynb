## Importing the required libraries

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

### Storing the given url in the String, and by using the request we access the data from the website

```python
#the url in which it contains website link
url = "https://www.amazon.in/s?rh=n%3A6612025031&fs=true&ref=lp_6612025031_sar"
r = requests.get(url)
r
```

```python
htmlContent = r.content
# htmlContent
```

### Creating BeautifulSoup object and parsing the HTML Content

```python
soup = BeautifulSoup(htmlContent,'html.parser')
s1 = soup.prettify()
# print(s1)
```

### Storing the data in the lists of names detailList, priceList, and ratingList

```python
details=""
detailList=[]
for details in soup.find_all('span','a-size-base-plus a-color-base a-text-normal',text=True):
    details=details.get_text()
    detailList.append(details)
detailList
```

```python
price=""
priceList=[]
for price in soup.find_all('span','a-price-whole'):
    price=price.get_text()
    priceList.append(price)
priceList
```

```python
ratings=""
ratingList=[]
for ratings in soup.find_all('span','a-icon-alt'):
    ratings=ratings.get_text()
    ratingList.append(ratings)
ratingList
```

### Adding the lists into the dataframe

```python
#Create a dataframe name df
df = pd.DataFrame([detailList,priceList,ratingList])
df
```

### Transposing the dataframe and storing it in a data variable

```python
data = df.transpose()
```

```python
data
```

### Renaming the Columns using rename() function

```python
#rename the columns
data = data.rename(columns={0:'Product',1:'Price',2:'Rating'})
data
```

### Creating Seller Name column

```python
li = pd.DataFrame(data['Product'].str.split(' ',1,expand=True))
li
```

```python
li = li.iloc[:,0]
data['SellerName'] = li
```

### Overwriting the Rating column

```python
ri = pd.DataFrame(data['Rating'].str.split(' ',1,expand=True))
ri = ri.iloc[:]
data["Rating"]= ri
data
```
